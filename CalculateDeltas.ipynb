```python
import pandas as pd
```

```python
startYear = 2017
endYear = 2019

dfs = {}

for year in range(startYear, endYear+1):
    dfs[year] = pd.read_csv("Data/playerStats_Standard_" + str(year) + ".tsv", sep="\t")
```

```python
def CleanData(df):
    # get rid of meaningless column
    df = df.drop(columns=["Rk"])
    
    # consolidate duplicates into single season
    for name in df.Player.unique():
        dupes_df = df.loc[df.Player == name]
        
        if len(dupes_df) > 1:
            #print(dupes_df)
            toDrop = list(dupes_df.index)[1:]
            #print(toDrop)
            df = df.drop(index=toDrop)
    
    df = df.reset_index(drop=True)
    
    return df
```

```python
# clean all data
for year in range(startYear, endYear+1):
    dfs[year] = CleanData(dfs[year])
    print(dfs[year])
```
