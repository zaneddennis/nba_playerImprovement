```python
import pandas as pd
import re
```

```python
def RemoveStar(playerName):
    return re.sub(r"\*", "", playerName)

def CleanDf(df):
    # remove blank rows
    
    
    #remove * from names
    df["Player"] = df["Player"].map(RemoveStar)
    
    return df
```

```python
startYear = 2000
endYear = 2019

dfs = {}

for year in range(startYear, endYear+1):
    tables = pd.read_html("https://www.basketball-reference.com/leagues/NBA_" + str(year) + "_per_game.html")
    dfs[year] = tables[0]
    dfs[year] = CleanDf(dfs[year])
```

```python
for year in range(startYear, endYear+1):
    print(year)
    print(dfs[year])
    dfs[year].to_csv("Data/playerStats_Standard_" + str(year) + ".tsv", sep="\t", index=False)
```
